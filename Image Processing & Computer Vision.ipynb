# Image Rotation

```python
height,width=img.shape[:2]
```

```python
rotationMatrix=cv2.getRotationMatrix2D((width/2,height/2),10,0.7)
```

```python
rotated_img=cv2.warpAffine(img,rotationMatrix,(width,height))
```

```python
plt.imshow(rotated_img)
plt.show()
```

# Resize Image

```python
img=cv2.imread('C:/R DATA SETS/Orange.jpg')
img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
newImg =cv2.resize(img, (0,0), fx=0.2, fy=0.75)
plt.imshow(newImg)
plt.show()
```

```python
newImg= cv2.resize(img, (1000,200))
plt.imshow(newImg)
plt.show()
```

# Image Cropping

```python
crop_image =img[100:500,300:700]
plt.imshow(crop_image)
plt.show()
```

# Image Cropping

```python
start_row, start_col = int(height * .25), int(width * .25)
end_row, end_col = int(height * .75), int(width * .75)
cropped = img[start_row:end_row , start_col:end_col]
```

```python
plt.imshow(cropped)
plt.show()
```

# Shapening

```python
plt.imshow(img)
plt.show()
```

```python
kernel_sharpening = np.array([[-1,-1,-1],
[-1,9,-1],
[-1,-1,-1]])
sharpened = cv2.filter2D(img, -1, kernel_sharpening)
```

```python
plt.imshow(sharpened)
plt.show()
```

# Blurring

```python
kernel1 = np.ones((3, 3)) / 10
blurred1 = cv2.filter2D(img, -1, kernel1)
```

```python
plt.imshow(blurred1)
plt.show()
```

```python
kernel2 = np.ones((7, 7)) / 100
blurred2 = cv2.filter2D(img, -1, kernel2)
```

```python
plt.imshow(blurred2)
plt.show()
```
